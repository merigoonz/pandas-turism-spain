## Data transformation

Original dataframe needs some modifications as we would like to extract specific data so at the beggining it is necessary to perform the following steps:
- Remove first column as it does not contains any important data as discovered in previous steps. 
- Remove rows where column 'Continente y país de residencia: Nivel 2' is null as I would like to know the country and continent from not nulls values
- As column 'Continente y país de residencia: Nivel 4' in some cases contains the country and on 'Continente y país de residencia: Nivel 3' column is stored the contient specification that is not going to be used in that exercice I will copy the data contained on column 'Continente y país de residencia: Nivel 3' where column 'Continente y país de residencia: Nivel 4' is null. 
- Remove column 'Continente y país de residencia: Nivel 3'
- It is necessary to split column 'Periodo' on two different columns, one will contains the year and the other the month number.
- Modify column description in order to have a clear dataframe

  Columns: 
  Continente y país de residencia: Nivel 2	-> Continent
  Continente y país de residencia: Nivel 4  -> Country
  CCAA de destino -> CCAA destination





Proceed to:
- Remove first column as it does not contains any important data as discovered in previous steps. 
- Remove rows where columns 'Continente y país de residencia: Nivel 2' and 'Continente y país de residencia: Nivel 3' are null as I would like to know the country and continent from not nulls values

```python
tourismData = (tourismData \
  .drop('Continente y país de residencia: Nivel 1', axis=1) \
  .dropna(subset=['Continente y país de residencia: Nivel 2', 'Continente y país de residencia: Nivel 3']) 
)
```

Perform step:
As column 'Continente y país de residencia: Nivel 4' in some cases contains the country and on 'Continente y país de residencia: Nivel 3' column is stored the contient specification that is not going to be used in that exercice I will copy the data contained on column a new column called 'Country' where column 'Continente y país de residencia: Nivel 4' is not null and when it is null I will copy 'Continente y país de residencia: Nivel 3'.

```python
tourismData['Country'] = tourismData['Continente y país de residencia: Nivel 4'].fillna(tourismData['Continente y país de residencia: Nivel 3'])
```

```python
tourismData.head(5)
```

Steps:
- Remove column 'Continente y país de residencia: Nivel 3'
- It is necessary to split column 'Periodo' on two different columns, one will contains the year and the other the month number.
- Modify column description in order to have a clear dataframe

```python
tourismData = (tourismData \
  .assign(
    year = lambda dataset: dataset["Periodo"].str.split("M").str.get(0),
    month = lambda dataset: dataset["Periodo"].str.split("M").str.get(1)
  ) \
  .rename(  columns = { "Continente y país de residencia: Nivel 2": "Continent", "CCAA de destino": "CCAA destination", "Concepto turístico": "Tourist concept"}) \
 .drop(['Continente y país de residencia: Nivel 3','Continente y país de residencia: Nivel 4','Periodo'], axis=1) \
)
```
